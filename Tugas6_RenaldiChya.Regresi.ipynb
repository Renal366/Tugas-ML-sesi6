```python
import numpy as np
import matplotlib.pyplot as plt


# Buat dataset sederhana
np.random.seed(0)  # Untuk hasil yang konsisten
X = 2 * np.random.rand(100, 1)  # Membuat 100 titik fitur antara 0-2
y = 4 + 3 * X + np.random.randn(100, 1)  # Membuat target dengan pola linear + noise

#Visualisasi dataset
plt.figure(figsize=(10, 6))
plt.scatter(X, y, alpha=0.7, color='blue')  # Plot titik-titik data
plt.xlabel('Fitur (X)', fontsize=12)
plt.ylabel('Target (y)', fontsize=12)
plt.title('Dataset Sederhana untuk Latihan Regresi Linear', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)

#Menambahkan garis regresi sebenarnya (hijau) untuk referensi
X_line = np.array([[0], [2]])
y_line = 4 + 3 * X_line
plt.plot(X_line, y_line, 'g-', linewidth=3, label='Hubungan Sebenarnya: y = 4 + 3X')

plt.legend(fontsize=12)
plt.show()


# Plot dataset
plt.scatter(X, y)
plt.xlabel('Fitur')
plt.ylabel('Target')
plt.title('Dataset Sederhana untuk Latihan Regresi')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Latih model
model = LinearRegression()
model.fit(X, y)

# Prediksi
X_new = np.array([[0], [2]])
y_pred = model.predict(X_new)

# Plot prediksi model
plt.plot(X_new, y_pred, 'r--', label='Prediksi Model')
plt.legend()
plt.show()

# Tampilkan parameter model
print(f"Intersep (bias): {model.intercept_[0]:.2f}")
print(f"Koefisien (slope): {model.coef_[0][0]:.2f}")
```
